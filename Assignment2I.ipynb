```python
#Task 2.3 
#What are the factors contributing to low and high income?
#Does the working class type of the individual matter in earning more income?

# Creating Pandas Series for the workclasses whose income is higher than 50K 
df_ = df1.loc[df1['income'] == '>50K',['workclass']]
workclass_types = df_['workclass'].value_counts()
labels = list(workclass_types.index)
aggregate = list(workclass_types)

# This Pie chat shows the Percentage of different workclass who earns more than 50K
plt.pie(aggregate, labels = labels, autopct='%1.2f%%', shadow=True)
plt.legend(labels, loc = 'best')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
# Changing the Categorical Values to Numerical values using the Label Encoder
from sklearn.preprocessing import LabelEncoder

categorical_features = list(df1.select_dtypes(include=['object']).columns)
label_encoder_feat = {}
for i, features in enumerate(categorical_features):
    label_encoder_feat[features] = LabelEncoder()
    df1[features] = label_encoder_feat[features].fit_transform(df1[features])

df1.head()
```

```python
#Task 3.1 

from sklearn.utils import shuffle
df1 = shuffle(df1)
from sklearn import preprocessing

# Splitting the data set into train and test set
from sklearn.model_selection import train_test_split


#TrainingSuite1
features = df1.drop(columns = ['income'])
target = df1['income']
X1_train, X1_test, Y1_train, Y1_test = train_test_split(features, target, test_size = 0.5,random_state = 0)

print ("Train data set size1 : ", X1_train.shape)
print ("Test data set size 1: ", X1_test.shape)



#TrainingSuite2
features = df1.drop(columns = ['age'])
target = df1['age']
X2_train, X2_test, Y2_train, Y2_test = train_test_split(features, target, test_size = 0.4,random_state = 0)

print ("Train data set size 2: ", X2_train.shape)
print ("Test data set size :2 ", X2_test.shape)



#TrainingSuite3
features = df1.drop(columns = ['workclass'])
target = df1['workclass']
X3_train, X3_test, Y3_train, Y3_test = train_test_split(features, target, test_size = 0.2,random_state = 0)

print ("Train data set size 3: ", X3_train.shape)
print ("Test data set size 3: ", X3_test.shape)
```

```python
#Task 3.2 Performing steps on 3 suites

from sklearn.linear_model import LogisticRegression
import numpy as np
from sklearn                        import metrics, svm
from sklearn.linear_model           import LogisticRegression
from sklearn import preprocessing
from sklearn import utils



#LogisticRegression
# Training the model_1
logistic = LogisticRegression(C = 0.5, max_iter = 500)
model_1 = logistic.fit(X1_train, Y1_train)

# Predictions
pred_1 = model_1.predict(X1_test)



# Training the model_2
logistic = LogisticRegression(C = 0.4, max_iter = 500)
model_2 = logistic.fit(X2_train, Y2_train)

# Predictions
pred_2 = model_2.predict(X2_test)



# Training the model_3
logistic = LogisticRegression(C = 0.5, max_iter = 500)
model_3 = logistic.fit(X3_train, Y3_train)

# Predictions
pred_3 = model_3.predict(X3_test)


```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix


lr_acc=accuracy_score(Y1_test, pred_1)
print ("The accuracy of model 1 : ",lr_acc)


#model 1 confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(Y1_test, pred_1)
print(confusion_matrix)


#Classificstionreport1
print classification_report(Y1_test,pred_1)



#Model 2
lr_acc2=accuracy_score(Y2_test, pred_2)
print ("The accuracy of model 2 : ",lr_acc2)



from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(Y2_test, pred_2)
print(confusion_matrix)

print classification_report(Y2_test,pred_2)



#Model 3
lr_acc3=accuracy_score(Y3_test, pred_3)
print ("The accuracy of model 3 : ",lr_acc3)

from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(Y3_test, pred_3)
print(confusion_matrix)

print classification_report(Y3_test,pred_3)

```

```python
#Task 3.3Comparison of Parameters 


from sklearn import model_selection 
from sklearn.metrics import accuracy_score


import sklearn.model_selection
sklearn.model_selection.KFold
    
from sklearn.metrics import accuracy_score
seed=7
models=[]
models.append(('RFC',RandomForestClassifier()))
models.append(('LR',LogisticRegression()))
results=[]
names=[]
scoring='accuracy'
for name,model in models:
    sklearn.model_selection.KFold
    kfold = model_selection.KFold(n_splits=10,random_state=seed)
    cv_results=model_selection.cross_val_score(model,X1_test,Y1_test, cv=kfold)
    results.append(cv_results)
    names.append(name)
msg= '%s: %f(%f)' %(names,cv_results.mean(),cv_results.std())
print(msg)
```

```python
#Comparison of accuracy for two models
fig=plt.figure()
fig.suptitle('Comparison of Algorithm')
plt.xlabel('models')
plt.ylabel('accuracy score')
ax=fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```
